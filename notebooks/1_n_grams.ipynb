```
from nltk import ngrams
```

```
# load our sample string / input
sentence = "The Philippines was named in honor of King Philip II of Spain. Spanish explorer Ruy López de Villalobos, during his expedition in 1542, named the islands of Leyte and Samar Felipinas after the then-Prince of Asturias. Eventually the name Las Islas Filipinas would be used to cover all the islands of the archipelago. Before that became commonplace, other names such as Islas del Poniente (Islands of the West) and Magellan's name for the islands San Lázaro were also used by the Spanish to refer to the islands"

# save the number of n, in this case. unigram
n = 1
```

```
# apply the ngrams class of nltk
unigrams = ngrams(sentence, n)

# print the output of unigram
print('Unigrams :{}'.format(unigrams))
```

```
for gram in unigrams:
   print(gram, end=' ')
```

```
# apply bigram
n = 2

bigrams = ngrams(sentence, n)

for gram in bigrams:
   print(gram, end=' ')
```

```
# apply trigram
n = 3

trigrams = ngrams(sentence, n)

for gram in trigrams:
   print(gram, end=' ')
```

```
# wordgram

#split the sentence
words = sentence.split()

print(words)
```

```
# loop thre iterations
for count in range(3):
  n = count + 1

  current_grams = ngrams(words, n)
  
  if(n == 1):
    print('Unigrams: ')
  elif(n == 2):
    print('\nBigrams: ')
  elif(n == 3):
    print('\nTrigrams: ')
  
  for gram in current_grams:
    print(gram, end=' ')
```

```
# save unigrams here
unigram_list = []

n = 1
unigrams = ngrams(corpus,n)

for gram in unigrams:
  unigram_list.append(gram)
```

```
from collections import defaultdict
```

```
dictionary = defaultdict(list)

# make a temporary value for the previous word
previous_word = ''

for current_word in unigram_list:
  if previous_word !='':
    dictionary[previous_word].append(current_word[0])
  previous_word = current_word[0]
```

```
dictionary
```

```
def compute_conditional_probabilities(corpus):
  for key in dictionary.keys():
    # grab next word from dict
    next_word = dictionary[key]
    
    unique_words = set(next_word)
    
    num_words = len(next_word)
    
    # placeholder dict for the loop
    probabilities_given_key = {}
    
    #loop for serving value of current key
    for unique_word in unique_words:
      # save tje value of the current word
      probabilities_given_key[unique_word] = float(next_word.count(unique_word)
                                                  ) / num_words
  return probabilities_given_key
```

```
compute_conditional_probabilities(dictionary)
```

```
# define a function for predicting the probability of 
def next_word_predictor(conditional_probabilities, current_word, next_word):
  # if current word is existng, check probability for the next word
  if current_word in conditional_probabilities:
    if next_word in conditional_probabilities[current_word]:
      return conditional_probabilities[current_word][next_word]
  return 0.0
   
```

```
conditional_probability = compute_conditional_probabilities(corpus)
```
